```java
//Importa a classe Scanner.
import java.util.Scanner;

//Inicia a classe Scanner para entrada de dados.
Scanner dados= new Scanner(System.in);

//Imprime a mensagem.
System.out.print("Insira o primeiro número: "); 
//Declara e inicia a variável arg1.
int arg1 = dados.nextInt();

//Imprime a mensagem.
System.out.print("Insira o segundo número: ");
//Declara e inicia a variável arg2.
int arg2 = dados.nextInt();

//Fecha o objeto dados da classe Scanner.
dados.close();

//Imprime a mensagem.
System.out.print("O resultado da soma é: ");

//Imprime a soma de arg1 e arg2.
System.out.println(arg1+arg2);
```
